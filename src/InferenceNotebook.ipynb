```python
from dualrnn_test_wav import Separation
from pathlib import Path
from random import choice

import IPython.display as ipd
import os
```

```python
my_separation = Separation("config/Dual_RNN/train.yaml", "checkpoints/Dual_Path_RNN/best.pt", [0])
```

```python
him_separation = Separation("config/Dual_RNN/train.yaml", "original.pt", [0])
```

```python
sr = 8000
```

```python
def separate_wav(separation, audio_path, separated_path="output"):  
    separation.inference(audio_path, separated_path)
    print("Original mix")
    ipd.display(ipd.Audio(audio_path, rate=sr))
    print("First speaker")
    ipd.display(ipd.Audio(Path(separated_path) / "spk1" / "test.wav" , rate=sr))
    print("Second speaker")
    ipd.display(ipd.Audio(Path(separated_path) / "spk2" / "test.wav" , rate=sr))
```

```python
data_path = Path('..') / 'data' / 'test'
```

```python
random_folder = choice(list(data_path.iterdir()))
```

```python
%%time
separate_wav(my_separation, random_folder / 'mix_audio.wav')
```

```python
separate_wav(him_separation, random_folder / 'mix_audio.wav')
```
