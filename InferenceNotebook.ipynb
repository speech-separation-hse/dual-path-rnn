```python
from dualrnn_test_wav import Separation
from pathlib import Path

import IPython.display as ipd
import os
```

```python
sr = 8000
```

```python
def separate_wav(audio_path, separated_path="output", preproc_path="out_wav.wav"):
    os.system(f"ffmpeg -i {audio_path} -ar 8000 -ac 1 {preproc_path} -y")
    separation = Separation(preproc_path, "config/Dual_RNN/train_rnn.yml", "best.pt", [0])
    separation.inference(separated_path)
    print("Original mix")
    ipd.display(ipd.Audio(preproc_path, rate=sr))
    print("First speaker")
    ipd.display(ipd.Audio(Path(separated_path) / "spk1" / "test.wav" , rate=sr))
    print("Second speaker")
    ipd.display(ipd.Audio(Path(separated_path) / "spk2" / "test.wav" , rate=sr))
```

```python
%%time

separate_wav("girls.wav")
```
